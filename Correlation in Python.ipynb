Load Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Data

```python
power_lifting = pd.read_csv(r'C:/Users/Hanna Bohannon/Documents/DSO105C/DSO105C Datasets/power_lifting.csv')
power_lifting.head()
```

Run One Correlation

```python
power_lifting['BestBenchKg'].corr(power_lifting['BodyweightKg'])
```

Drop Non-Continuous Variables

```python
power_lifting1 = power_lifting.drop(['MeetID', 'Name'], axis=1)
power_lifting1.head()
```

Using .corr() on An Entire Dataset

```python
power_lifting1.corr(method='pearson')
```

Make .corr() Pretty!

```python
power_lifting1.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

Use sns.heatmap()

```python
sns.heatmap(power_lifting1.corr(), annot=True)
```
